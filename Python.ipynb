```python
def decorated_user(funcp):
    def greet_user():
        print("Have a nice day", end=" ")
        funcp()
    return greet_user
    
@decorated_user
def username():
    name="vaishnavi"
    print(f"{name}")

username()
username()
```

```python
def divide_decorator(funcp):
    def custom_divide(v1,v2):
        if v2==0:
            print("Cannot divide by zero")
            return
        return funcp(v1,v2)
    return custom_divide

@divide_decorator
def divide(num,den):
    return num/den

print(divide(10,3))
print(divide(10,0))
```

```python
def decorator1(ref):
    def decor1():
        return ref()*2
    return decor1

def decorator2(ref):
    def decor2():
        return ref()*10
    return decor2

@decorator2
@decorator1
def num():
    return 10

print(num())
```

```python
def add(*args):
    total=0
    for value in args:
        for eachVal in value:
            total+=eachVal
    return total

def check(*args):
    return add(args)

print(add([1,2,3,4,5]))
print(add([15, 20, 25, 30]))
#print(add(10,20,30))
print(check(1,2,3,4,5))
```
